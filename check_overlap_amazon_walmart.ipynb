```python
import sys
import os
from pprint import pprint
import pandas as pd
import csv
import math
```

```python
isbns = []
df_walmart = pd.read_csv("final_books_csv_data/walmart_books.csv", encoding='cp1252')
walmart_isbn = df_walmart['ISBN10'].tolist()
df_amazon = pd.read_csv("final_books_csv_data/amazon_books.csv", encoding='cp1252')
amazon_isbn = df_amazon['ISBN10'].tolist()
for isbn in walmart_isbn:
    if isbn in amazon_isbn and not pd.isnull(isbn):
        isbns.append(isbn)
        print(isbn)
print(len(isbns))
print(len(set(isbns)))
```
